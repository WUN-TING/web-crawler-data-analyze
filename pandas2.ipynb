```python
import requests
import pandas as pd

aqi_url = "http://tbike-data.tainan.gov.tw/Service/StationUseageStatics/json"
r = requests.get(aqi_url)
print(r.text)
```

```python
aqi = pd.read_json(r.text)
print(type(aqi))
```

```python
# MUTATE
import numpy as np

def get_na(x):
    if x=="" or x== "ND":
        return np.nan
    else:
        return x
    
```

```python
aqi["場站"] = aqi["場站"].apply(get_na)
```
